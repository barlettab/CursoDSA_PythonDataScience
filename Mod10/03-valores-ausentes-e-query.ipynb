### Preenchendo Valores ausnetes em DataFrames do Pandas

A função `fillna()`é usada para preencher os valores ausentes. A função oferece muitas opções. Podemos usar um valor específico, uma função agregada (por exemplo, a média) ou um valor anterior ou seguinte. No exemplo abaixo, usaremos a moda para preencher o NaN.

```python
import pandas as pd
import numpy as np
```

```python
# Importanto o dataset
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
# Somando a quantidade de valores ausentes de cada coluna
df.isna().sum()
```

```python
# Extraímos a moda da coluna Quantity
mode = df['Quantidade'].value_counts().index[0]
```

A __moda__ em Estatística é uma medida de tendência central que representa o valor mais frequente em um conjunto de dados. 

A moda é especialmente útil quando queremos saber qual é o valor mais comum ou popular em um conjunto de dados, seja em uma distribuição unimodal ou em uma distribuição bimodal. 

No entanto, a moda pode não ser representativa quanto outras medidas de tendência central, como a média e a mediana, especialmente em distribuições assimétricas ou quando há valores extremos (outliers). Por essa razão é importante analisar diferentes medidas de tendência central e usar a que melhor se adequa aos objetivos da análise estatística.

```python
# Imprimindo o valor da moda (o valor que apareceu mais vezes)
print(mode)
```

```python
# Preenchendo os valores NaN da coluna Quantidade com a moda
df.fillna({'Quantidade': mode}, inplace = True)
```

```python
# Verificando se a quantidade de NaN
df.isna().sum()
```

`Observação:` Usar o `inplace=True` garante que a alteraçãp seja feita no dataset original e não em uma cópia do dataframe.

### Query de Dados no DataFrame do Pandas
Com o Pandas criamos dataframes que são essencialmente tabelas. Como tal, podemos fazer consultas, ou simplesmentes queries. E para isso usamos o método `query()`.

```python
df.head()
```

```python
# Checamos os valores mínimo e máximo da coluna Valor_Venda
df.Valor_Venda.describe()
```

O intervalo de vendas __Valor_Venda__ é de 0,44 a 22638. Vamos fazer uma consulta e retornar as vendas dentro de um range de valores.

```python
# Geramos um novo df aapeas com o intervalor de vendas entre 229 e 10000
df2 = df.query('229 < Valor_Venda < 10000')
```

```python
df2.Valor_Venda.describe()
```

```python
# Geramos um novo df apenas com os valores acima da média
df3 = df2.query('Valor_Venda > 766')
```

```python
df3.head()
```

### Verificando a Ocorrência de Diversos Valores em Uma Coluna

Em nosso conjunto de dados do exemplo temos a coluna Quantidade que representa a quantidade de itens vendidos em cada uma das vendas. Imagine que precisamos saber em quais vendas foram vendidos 5, 7, 9 ou 11 itens. O Pandas oferece o método `isin()`para checar diversos valores em uma coluna. Esse método tem a mesma função do __IN__ presente na linguagem SQL.

```python
# Retornando o número de linhas e colunas
df.shape
```

```python
df[df['Quantidade'].isin([5, 7, 9, 11])]
```

```python
# Shape 
df[df['Quantidade'].isin([5, 7, 9, 11])].shape
```

```python
# Mostrando as 10 primeiras linhas
df[df['Quantidade'].isin([5, 7, 9, 11])][:10]
```

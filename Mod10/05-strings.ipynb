### Filtrando DataFrame do Pandas com Base em Strings

O Pandas oferece diversas funções para manipulação de strings. Começaremos com os filtros de strings com base nas letras iniciais e finais.

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
# Filtramos os dataframe pela coluna Segmento com valores que iniciam com as letras 'Con'

df[df.Segmento.str.startswith('Con')].head()
```

```python
df.Segmento.value_counts()
```

```python
# Filtramos os dataframe pela coluna Segmento com valores que terminam com as letras 'mer'

df[df.Segmento.str.endswith('mer')].head()
```

`Observação:` As funções `startswith() e `endswith()` são muito úteis quando for necessário filtrar strings por caracteres que apareçam no início ou no final.

### Split de Strings em DataFrames do Pandas


```python
df['ID_Pedido'].head()
```

O formato dos dados da coluna "ID_Pedido" apresenta o país, o ano e o id do pedido. Esse formato não é o ideal para a manipulação de dados. Então, vamos dividir essa coluna e extrair o ano, gravando-o em um nova coluna.

```python
# Split da coluna pelo caracter '-'
df['ID_Pedido'].str.split('-')
```

O resultado são listas em Python. Para extrair o ano precisamos especificar o índice da posição que queremos extrair (em nosso caso a posição 2, logo, índice 1 em Python):

```python
df['ID_Pedido'].str.split('-').str[1].head()
```

```python
# Criando a coluna ano com base no split
df['Ano'] = df['ID_Pedido'].str.split('-').str[1]
```

```python
df.head()
```

### Strip de Strings em DataFrames do Pandas
O `split` divide a string, o `strip` remove caracteres da string.

```python
df['Data_Pedido'].head()
```

A coluna 'Data_Pedido' é a data de envio do produto no formato YYYY-MM-DD. Imagine que seja necessário deixar o ano apenas com 2 dígitos sem alterar o tipo da variável. Fazemos isso com a função `lstrip()`, ou seja, _left strip_.

```python
# Removendo os dígitos 2 e 0 à esquerda do valor da variável 'Data_Pedido'
df['Data_Pedido'].str.lstrip('20')
```

`Observação:` Como não usamos o `inplace=True`a mudança é somente na memória e não altera o dataframe. Além disso, podemos usar as funções `rstrip()`e `strip()`.

### Replace de Strings em DataFrames 

```python
df['ID_Cliente'] = df['ID_Cliente'].str.replace('CG', 'AX')
```
